##  API usage
* #### Upload data, tags and compute model
#### `/admin/computeModelFromCsvWithTags`
##### Parameters
* ##### `data_file` - csv file with implicit data.
* ##### `tags_file` - json file with items features.
##### Output
* ##### If successful: _Sucesfully computed recommender model_.
* ##### Otherwise: _Message with error_.



##  API usage
* #### Get recommendations for user
#### `/recommendation/<userName>`
##### Parameters
* ##### None
##### Output
* ##### List of top 20 recommended presentations.
* ##### List of rank for each presentation.
* ##### Username.
